```python
import math
def concentracio ( Co, k ,t):
    concentracio = []
    for i in range (t): 
        C = Co*math.exp(-k*t)
        concentracio.append(C)
    return concentracio


Co = 200
k= 3
t= 25


print(concentracio (Co,k,t))
```

```python
def analitza_llista ( llista, element):
    comptador = 0

    for i in llista:
         if i == element:
            comptador += 1
            
    if comptador > 0:            
        return True, comptador
    else: 
        return False
            
llista = [2,6,4,7,8,5,3]        
element = 2

print(analitza_llista ( llista, element))
```
